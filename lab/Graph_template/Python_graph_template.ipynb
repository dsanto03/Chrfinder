```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

cwd = Path('.')
print(cwd.resolve())

### Import all values from a file to make a table named "values"
### Just change the path, the names and if need the skiprows

file_path = Path(cwd / "Graph_template" / 'Values_template.txt')
values = pd.read_csv(file_path, delimiter = '\t',names=["Pressure [bar]", "Flux [L/min]", "U [V]", "I [A]", "P [W]", "Temp. [Â°C]"], skiprows=1)

values = values.apply(lambda x: x.str.replace(',','.').astype(float))  # replace all ',' by '.'

#print(values)


### Make a graphic of the values

fig, ax = plt.subplots()  # creat a figure and a variable that take axis
ax.plot(values["I [A]"], values["U [V]"], linestyle="-", color='black', zorder=2)  # plot a black line through all values
ax.scatter(values["I [A]"].iloc[0:34], values["U [V]"].iloc[0:34], color='#1f77b4', zorder=3, label='$\eta_{conc}$', s=45)  # change the color of a set of points and add a label to the legend
ax.scatter(values["I [A]"].iloc[34:], values["U [V]"].iloc[34:], color='purple', zorder=3, label=r'Linear function: $\eta_{ohm}$', s=45)  # change the color of a set of points and add a label to the legend


# ax.set_xlim(min_value_x, max_value_x)  # Adjust min_value_x and max_value_x according to your data
# ax.set_ylim(min_value_y, max_value_y)  # Adjust min_value_y and max_value_y according to your data


ax.set_xlabel("Intensity [A]", fontsize='14')  # label the x axis
ax.set_ylabel("Tension [V]", fontsize='14')  #label the y axis
ax.tick_params(axis='both', which='major', labelsize=12)  # Adjust labelsize as needed

legend = ax.legend(title='Predominant effect', loc='upper right', fontsize='12')  # Set the title and position of the legend, also size of the legend labels
legend.set_title('Predominant effect', prop={'size': 12})  # Set the title font size
legend.get_frame().set_facecolor('#e0e0e0')  # Set the background color of the legend
legend.get_frame().set_edgecolor('black')  # Set the edge color of the legend frame


### save
#plt.savefig('tension_vs_intensity.png')
```
