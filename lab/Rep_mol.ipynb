```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG=True  #using Scalable Vector Graphics ensure a better image quality as PNG or JPEG.

# 2D structure representations from smiles.

molecule_smiles = "C12=C3C4=C5C6=C1C7=C8C9=C1C%10=C%11C(=C29)C3=C2C3=C4C4=C5C5=C9C6=C7C6=C7C8=C1C1=C8C%10=C%10C%11=C2C2=C3C3=C4C4=C5C5=C%11C%12=C(C6=C95)C7=C1C1=C%12C5=C%11C4=C3C3=C5C(=C81)C%10=C23"
mol_2D = Chem.MolFromSmiles(molecule_smiles)
mol_2D
```

```python
# molecule_smiles = "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"
# mol_2D = Chem.MolFromSmiles(molecule_smiles)

# 3D structures from smiles.

from rdkit.Chem import AllChem

# Add H to the molecule
mol_H = Chem.AddHs(mol_2D) 
# Create SDF format of the molecule
AllChem.EmbedMolecule(mol_H, AllChem.ETKDG())  # SDF mol format: table containing all the coordinates of each atom in 3D
# print(Chem.MolToMolBlock(mol_H))  

import py3Dmol

# Helpful function from a blog (https://rdkit.blogspot.com/2016/07/a-recent-post-on-in-pipeline-talked.html)
def drawit(m,p=None,confId=-1,sphere_radius=0.3):
        mb = Chem.MolToMolBlock(m,confId=confId)
        if p is None:
            p = py3Dmol.view(width=400,height=400)
        p.removeAllModels()
        p.addModel(mb,'sdf')
        p.setStyle({'stick':{'radius': sphere_radius}}) # use: "sphere" for filled space representation. 
        p.setBackgroundColor('0xeeeeee')
        p.zoomTo()
        return p.show()

drawit(mol_H)   # print in 3D the SDF format representation.

```
